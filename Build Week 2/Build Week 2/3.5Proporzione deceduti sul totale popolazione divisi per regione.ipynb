```python
#importazione file pulito
df_regioni=pd.read_csv("covid19_region_python_w_daily_deaths.csv", sep = ",")
```

```python
#importazione del file
df_comuni = pd.read_csv("comuni_clean.csv",sep=";")
```

```python
#totale morti divisi per regione (somma)
deceduti_regione = df_regioni.groupby("RegionName")["Deaths"].sum()
```

```python
tot_popolazione = df_comuni["Popolazione2011"].sum()
```

```python
#tot morti/tot poplazione
proporzione = (deceduti_regione/tot_popolazione)*100
```

```python
proporzione.plot(kind="bar",colormap='plasma')
plt.title('Proporzione deceduti sul totale popolazione divisi per regione')
```
