```python
#import del csv preso in esame
df_province = pd.read_csv("covid19_province _python.csv", sep=";")
```

```python
# Casi Totali dal 6 dicembre 2020
data_6_dicembre = df_province[df_province['Date'].str.startswith('2020-12-06')]

# Ordinamento dei dati in base ai casi totali positivi
province_ordinate = data_6_dicembre.sort_values(by='TotalPositiveCases', ascending=False)

# Sorted per Nome e Casi
province_ordinate = province_ordinate[['ProvinceName', 'TotalPositiveCases']]

province_ordinate.head(25)  #TOP 25
```
