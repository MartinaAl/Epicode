```python
df_ripartizioni_reg=pd.read_csv("covid19_reg_ripartite.csv", sep=";")
```

```python
df_ripartizioni_reg['Date'] = pd.to_datetime(df_ripartizioni_reg['Date'])
```

```python
def stagioni(x):
    if x['Date'] <= pd.Timestamp(x['Date'].year, 3, 21, 18, 0):
        return 'Inverno'
    elif pd.Timestamp(x['Date'].year, 3, 22, 18, 0) <= x['Date'] <= pd.Timestamp(x['Date'].year, 5, 21, 18, 0):
        return 'Primavera'
    elif pd.Timestamp(x['Date'].year, 5, 22, 18, 0) <= x['Date'] <= pd.Timestamp(x['Date'].year, 9, 21, 18, 0):
        return 'Estate'
    else:
        return 'Autunno'



df_ripartizioni_reg['Stagioni'] = df_ripartizioni_reg.apply(stagioni, axis=1)

#si deve fare il conteggio
season_counts = df_ripartizioni_reg['Stagioni'].value_counts()

#Grafico a barre
plt.figure(figsize=(8, 6))
season_counts.plot(kind='bar', color='skyblue')
plt.title('Morti distribuiti in stagioni')
plt.xlabel('Stagione')
plt.ylabel('Numero di occorrenze')
plt.show()
```
