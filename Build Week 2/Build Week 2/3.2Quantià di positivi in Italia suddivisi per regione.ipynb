```python
df_regioni=pd.read_csv("covid19_region _python.csv", sep = ";")
```

```python
#somma dei casi positivi raggruppati per regione
positivi_regione = df_regioni.groupby(['RegionName'])["TotalPositiveCases"].sum()
```

```python
#grafico a barre --> quantià di positivi in Italia suddivisi per regione
positivi_regione.plot.bar()
plt.title("Quantià di positivi in Italia suddivisi per regione")
```
