```python
#import del dataset di interesse
df_regioni = pd.read_csv("covid19_region_python_w_daily_deaths.csv")
```

```python
#seleziono le colonne di interesse
df_regioni = df_regioni.iloc[:,[3, 4, 5, 6, 7, 8, 9, 11, 12]]
```

```python
# Eseguo correlazioni sulle varibiali di interesse
corr= df_regioni.corr()
```

```python
#creazione della heatmap
plt.figure(figsize=(18, 10))
plt.subplot(121)
sns.heatmap(corr, cmap='magma', annot=True)
plt.xticks(rotation = 90)
plt.title('Correlazione tra Variabili del dataset regioni')
#plt.savefig("Correlazione.png", dpi=300)
# plt.show()
```
