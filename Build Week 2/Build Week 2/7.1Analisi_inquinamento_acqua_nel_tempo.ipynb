```python
#importazione file fiumi
df_fiumi=pd.read_csv("fiumi_clean.csv", sep = ";")
```

```python
#inserimento  dello zero al posto dei nan in modo tale che si possono sommare i valori (anche quelli nulli)
df_fiumi = df_fiumi.fillna(0)
```

```python
df_fiumi['Regioni'] = df_fiumi.iloc[:,1].replace('Friuli-Venezia Giulia', 'Friuli Venezia Giulia')
df_fiumi['Regioni'] = df_fiumi.iloc[:,1].replace('Trentino-Alto Adige', 'Trentino Alto Adige')
```

```python
#rename della colonna regioni
df_fiumi = df_fiumi.rename(columns={'Regioni_2019': 'Regioni'})
```

```python
#cancello colonne superflue
#df_fiumi=df_fiumi.drop(df_fiumi.loc[:, 9:10].columns, axis = 1)
df_fiumi.drop(df_fiumi.iloc[:, 8:11], inplace = True, axis = 1)
```

```python
andamento_casi_regione = df_fiumi[['Macrobenthos_2019','Macrobenthos_2020',	 'Pesci_2019', 'Pesci_2020',	'Diatomee_2019','Diatomee_2020']].sum()
andamento_casi_regione.plot(kind="bar")
```

Prendendo in considerazione l'anno di seguito, e il 2020 è l'anno in cui si sono registrati maggiori casi positivi, si vede come rispetto all'altro anno, la qualità dell'acqua si abbassa
